Q2. Write a python program to print Python Exception Hierarchy.

```python
def print_exception_hierarchy(exception_class, indent=0):
    print(" " * indent + exception_class.__name__)
    for subclass in exception_class.__subclasses__():
        print_exception_hierarchy(subclass, indent + 4)

print_exception_hierarchy(BaseException)
```

Q4. Why LookupError class is used? Explain with an example KeyError and IndexError.

The LookupError class in Python is the base class for exceptions that occur when a lookup or indexing operation fails. 

```python
#key
import logging
logging.basicConfig(filename="key.log",level=logging.INFO)
my_dict = {'a': 1, 'b': 2, 'c': 3}

try:
    value = my_dict['d']
except KeyError as e:
    logging.error("Error: Key not found in the dictionary{}".format(e))
```

```python
#index
my_list = [1, 2, 3]

try:
    value = my_list[5] 
except IndexError:
    print("Error: Index out of range!")
```

Q5. Explain ImportError. What is ModuleNotFoundError?

```python
try:
    import non_existent_module
except ImportError:
    print("Error: Failed to import module!")
```
